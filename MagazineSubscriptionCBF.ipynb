```python
import json
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Load in the data
metaPath = 'Data/meta_Magazine_Subscriptions.json'
reviewPath = 'Data/Magazine_Subscriptions.json'

metaData = []
reviewData = []

for line in open(metaPath, 'r'):
    metaData.append(json.loads(line))
    
for line in open(reviewPath, 'r'):
    reviewData.append(json.loads(line))
```

```python
# convert metadata into pandas dataframe

metaDF = pd.DataFrame.from_dict(metaData)
metaDF.head()
```

```python
# convert reviews data into pandas dataframe

reviewDF = pd.DataFrame.from_dict(reviewData)
reviewDF.head()
```

```python
# Drop duplicates

#convert dataframe values to string in order to use drop_duplicates method
metaDF = metaDF.astype(str)
reviewDF = reviewDF.astype(str)

metaDF = metaDF.drop_duplicates()
reviewDF = reviewDF.drop_duplicates()
```

```python
# Look at the columns in metaDF
metaDF.info()
```

```python
# Look at sample cell values in metaDF 
pd.set_option('display.max_colwidth', None)
metaDF.iloc[[0]]
```

From the metadata, I will extract the following features:
- category
- description
- brand

```python
# Exploring the types of values in the feature
metaDF[['category']].value_counts()
```

```python
metaDF[['description']].value_counts()
```

We see that there are missing values filled in as '[]' in both the category and description column.  This has to be cleaned up.

```python
metaDF[['brand']].value_counts()
```

```python
# Replace rows with values '[]' to an empty string
features = ['category', 'description', 'brand']

for feature in features:
    metaDF[feature] = metaDF[feature].apply(lambda x: '' if x == '[]' else x)
```

```python
pd.set_option('display.max_colwidth', 20)

# Combining relevant features into a single feature

def combined_features(row):
    combined_features = ''
    for feature in features:
        combined_features = combined_features + ' ' + row[feature] 
    return combined_features

metaDF['combined_features'] = metaDF.apply(combined_features, axis = 1)
metaDF.head()
```

```python
# Extracting features 
cv = CountVectorizer()
count_matrix = cv.fit_transform(metaDF['combined_features'])
print("Count Matrix:", count_matrix.toarray())
```

```python
#cosine similarity matrix
cosine_sim = cosine_similarity(count_matrix)
pd.DataFrame(cosine_sim)
```

We can see that the cosine similarity scores on the diagonals are 1, this means that product 0 is 100% similar to itself.  We can see that product 0 is more similar to product 4 (cosine similarity score: 0.3892) than to product 3 (cosine similarity score: 0.2231). 

```python
# Inputting into the model
# Select an arbitrary product asin to test
magazine_user_likes = 'B00005N7NQ'

def get_index_from_asin(asin):
    # Get the index of the product from it's asin
    return metaDF[metaDF.asin == asin].index.values[0]

magazine_index = get_index_from_asin(magazine_user_likes)
```

```python
# Generate similar magazines matrix
similar_magazines = list(enumerate(cosine_sim[magazine_index]))
```

```python
#sort similar magazines list in descending order
sorted_similar_magazines = sorted(similar_magazines, key=lambda x:x[1], reverse=True)
pd.DataFrame(sorted_similar_magazines)
```

```python
# print categories associated with the chosen input magazine
metaDF[metaDF['asin'] == 'B00005N7NQ']['category'][0]
```

```python
# Print top 15 most similar magazines

pd.set_option('display.max_colwidth', None)


def get_magazine_from_index(index):
    return metaDF.iloc[index].loc['asin'], metaDF.iloc[index].loc['category']

i = 0
for magazine in sorted_similar_magazines:
    print(get_magazine_from_index(magazine[0]))
    i += 1
    if i > 15:
        break
```

We can see that the recommended magazines are all categorized as 'Professional & Educational Journals' and 'Professional & Trade.'  This is the same as the input magazine we fed into the model.  We can see that these recommendations make sense.
